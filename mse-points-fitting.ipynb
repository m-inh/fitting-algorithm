```
import math
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```
class Point2D:
  """Represent a 2-dimentional point in Euclidean space

  Attributes
  ----------
  x, y: float

  """
  def __init__(self, x, y):
    self.x = x
    self.y = y
  
  def __str__(self):
    return "(" + str(self.x) + ", " + str(self.y) + ")"
```

```
def get_center_point(ps):
  sumx = 0
  sumy = 0
  length = len(ps)
  for p in ps:
    sumx += p.x
    sumy += p.y

  return Point2D(sumx/length, sumy/length)

# Visualize sample data
def visualize_sample(ps):
  visualize_all(ps, None)


def visualize_all(ps, pc):
  plt.ylabel('Y')
  plt.xlabel('X')
  plt.axis('equal')

  # Draw points
  for i in range(len(ps)):
    p = ps[i]
    plt.scatter(p.x, p.y, color='r')

  # Draw center point
  if (pc is not None):
    plt.scatter(pc.x, pc.y, color='g')

  # plt.xlabel("step: " + str(iter) + ", threshold: " + str(threshold) + ", inliers: " + str(current_score))
  plt.show()
```

```
# Prepare sample data
n = 100

random_circles = datasets.make_circles(n_samples=n, shuffle=True, noise=0.1, random_state=None, factor=0.8)
random_points = random_circles[0]
pxs = []
pys = []
ps = []
pc = None

for i in range(n):
  x = random_points[i][0] * 1000
  y = random_points[i][1] * 1000
  ps.append(Point2D(x, y))
  pxs.append(x)
  pys.append(y)

# double data for testing
for i in range(n):
  x = random_points[i][0] * 1200
  y = random_points[i][1] * 1000
  ps.append(Point2D(x, y))
  pxs.append(x)
  pys.append(y)


visualize_sample(ps)
```

```
pc = get_center_point(ps)

visualize_all(ps, pc)
```
